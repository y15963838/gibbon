# Gibbion Console 端的命令演示

```python
import gibbon
```

- 新建一个 getter 用以发送请求与承载信息

```python
getter = gibbon.Amap()
```

### 查询方法一
1. 设置 getter 的搜索关键词、搜索城市（默认为城市为：广州，关键词为：地铁站）

```python
getter.set_keywords('医院')
```

```python
getter.set_city('长沙')
```

2. 拿取数据并存入 pandas DataFrame

```python
getter.get_datas()
```

3. 查看拿取到的数据

```python
getter.dataframe
```

4. 将数据导出为 csv 文件

```python
path = r'C:\work\export_data\长沙医院查询结果.csv'
```

```python
getter.export(path)
```
